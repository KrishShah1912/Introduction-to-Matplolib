# Matplotlib example workflow

```python
# Importing matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# Setup plot
fig, ax = plt.subplots(figsize=(5,5))

# Plot data
ax.plot(x,y)

# Customize plot
ax.set(title = "Simple Plot", xlabel = "x-axis", ylabel ="y-axis")

# Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show()
```

# Two options for subplots

```python
# Subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
# Subplot option 2
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
plt.show()
```

# Which one to use ? (pyplot method or matplotlib OO method)

* when plotting something quickly , okay to use the pyplot method 
* when plotting something advance , use OO method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter", x="age", y ="chol", c = "target", figsize =(10,6))
plt.show()
```

```python
# OO method (object-oriented method)
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = "scatter" , x="age" , y = "chol" , c = "target", ax = ax)
ax.set_xlim([45,100])
plt.show()
```

```python
# OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"] , y =over_50["chol"] , c = over_50["target"] )

# customize the plot
ax.set(title = "Hear-Disease-Cholestrol-Levels", xlabel = "Age" , ylabel = "Cholestrol level")

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = "--")
plt.show()
```

```python
over_50.head()
```

```python
# subplot of age, chol, thalach
```

```python
%matplotlib inline
fig, (ax0,ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10))

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"] , y = over_50["chol"] , c = over_50["target"])
# customize the ax0 plot
ax0.set(title = "Heart Disease Age & Cholestrol " , xlabel = "Age" , ylabel = "Cholestrol Level")
# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title  = "Target")
# Add horizonatal line to ax0
ax0.axhline(over_50["chol"].mean(), linestyle = "--")

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"] , y = over_50["thalach"] , c = over_50["target"])
# customize the ax1 plot
ax1.set(title = "Heart Disease Thalach & Cholestrol" , xlabel = "Max Hear Rate" , ylabel = "Cholestrol Level ")
# Add legen to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")
# Add horizontal line to ax1
ax1.axhline(over_50["thalach"].mean() , linestyle ="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis" , fontsize =16 , fontweight= "bold")

plt.show()
```

### This Plot display the information about the heart disease dataset....

```python
# Customizing the y and x axis limitation

%matplotlib inline
fig, (ax0,ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10))

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"] , y = over_50["chol"] , c = over_50["target"], cmap = "winter")
# customize the ax0 plot
ax0.set(title = "Heart Disease Age & Cholestrol " , xlabel = "Age" , ylabel = "Cholestrol Level")
# Change the x axis limits
ax0.set_xlim([50,80])
# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title  = "Target")
# Add horizonatal line to ax0
ax0.axhline(over_50["chol"].mean(), linestyle = "--")

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"] , y = over_50["thalach"] , c = over_50["target"] , cmap = "winter")
# customize the ax1 plot
ax1.set(title = "Heart Disease Thalach & Cholestrol" , xlabel = "Max Hear Rate" , ylabel = "Cholestrol Level ")
# Change the x axis limits
ax1.set_xlim([50,80])
# Change the y axis limts
ax1.set_ylim([60,200])
# Add legen to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")
# Add horizontal line to ax1
ax1.axhline(over_50["thalach"].mean() , linestyle ="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis" , fontsize =16 , fontweight= "bold")

plt.show()
```
